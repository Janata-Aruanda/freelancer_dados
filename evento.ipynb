```python
#Criando o banco de dados
conn = sqlite3.connect("evento.db")
print("Conexão com sucesso!")

#Crianco a tabela de dados
conn.execute('''
CREATE TABLE IF NOT EXISTS evento(
  id INTEGER PRIMARY KEY,
  nome_evento VARCHAR(100),
  data_evento VARCHAR(15),
  localizacao VARCHAR(100),
  convidados INTEGER,
  custo REAL(7,2)
);
''')

conn.commit()
print("Tabela criado com sucesso!")
```

# *DEPLOYAR *

```python
#importação do modulo
import pandas as pd
print("Importação com sucesso!")
```

```python
#abrindo a conexação:
conn = sqlite3.connect("evento.db")
print("conexão com sucesso!")

#jogando a comando sql para uma variavel
base = '''
SELECT * FROM evento
'''

#lendo uma query e jogando para uma variavel para criar uma datafarme
df = pd.read_sql_query(base, conn)
print("criação do datafarme com sucesso!")

#Fechar a conexão com o banco
conn.close()
print("Conexão fechado com suesso!")
```

```python
#visualizar o tipe da variavel
type(df)
```

```python
#exibindo o datafarme
display(df)
```

```python
#imprimindo a quantidade do datafarme 
col = len(df.columns)
row = len(df.index) 
print("Tamanho do datafarme\n" )
print(f"Coluna: {col}\nlinhas: {row}")
```

```python
#Acessando apenas uma linha do datafarme
#usando o metedo loc, ele puxa como fosse arry
printdf.loc[0]
```
